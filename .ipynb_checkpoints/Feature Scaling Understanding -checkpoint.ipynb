```python
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import r2_score

from scipy import stats
from scipy.stats import norm,skew
import math 
```

```python
data = load_boston() 

df_Boston = pd.DataFrame(data=data['data'], columns=data['feature_names'])
df_Boston['Price'] = data['target']

extract_df = df_Boston.loc[:, ['RM','Price']]
```

We know that RM and Price are highly correlated, now let's perform a simple train/test/split and linear regression and see how our model generalizes !

```python
f,(ax1,ax2) = plt.subplots(1,2,figsize=(15,5))
x = extract_df['RM'].values.reshape(-1,1)
y = extract_df['Price'].values.reshape(-1,1)
clf = LinearRegression()
#Train, test & split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)

clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
error = mean_squared_error(y_test,y_pred)
print(error)
```

Let's try and understand the distribution of Price and see if we can fix any skewness !

```python
log_price = pd.Series(np.log(extract_df['Price']))
f,(ax1,ax2) = plt.subplots(1,2,figsize=(15,10))
sns.distplot(log_price, ax=ax2,fit=norm,label='Log Price')
sns.distplot(extract_df['Price'],ax=ax1,fit=norm,label='Price')
plt.legend(loc='upper left')
plt.show()
```

```python
# Let's look at the probability plot of Log Price to ensure we've got a somewhat normal distribution. 
res = stats.probplot(log_price,plot=plt)
plt.show()
```

```python
# Log Transformations
x_trainlog = np.log(x_train)
y_trainlog = np.log(y_train)
x_testlog = np.log(x_test)
```

```python
# Performing log transformation on just Price & fitting Linear Model
clf.fit(x_train,y_trainlog)
y_pred = clf.predict(x_test)
y_pred = np.apply_along_axis(lambda x:math.exp(x),1,y_pred)
error = mean_squared_error(y_test,y_pred) 
print(error)
```

```python
# Performing log transformation on RM & Price & fitting Linear Model
clf.fit(x_trainlog,y_trainlog)
y_pred = clf.predict(x_testlog)
y_pred = np.apply_along_axis(lambda x:math.exp(x),1,y_pred)
error = mean_squared_error(y_test,y_pred)
print(error)
```

Looks like the best model was Log Price vs. RM (with an MSE of 34.76) !

Log transformations help in reducing the skewness of your distribution and helps residuals to have a more normal distribution. Now, let's try and add another predictor variable and see if we can get the MSE down ! Let's start by doing a correlation matrix to see what variables are highly correlated with Price.   


```python
# Correlation Matrix - Let's get the top 5 correlations for Price
k = 5
corrmat = df_Boston.corr()
cols=corrmat.nlargest(k,'Price')['Price'].index
cm = np.corrcoef(df_Boston[cols].values.T)
sns.heatmap(cm,xticklabels=cols.values,yticklabels=cols.values,annot=True,vmin=0,vmax=1)
plt.show()
```

Now, let's take all RM, ZN,B and DIS as my features and perform a scatter matrix. It does look like DIS and ZN are highly correlated. Let's investigate this further. 

```python
imp_features = df_Boston.loc[:, ['RM','ZN','B','DIS']]
sns.pairplot(imp_features)
plt.show()
```

It's hard to tell whether it's a highly correlated variable or not. I'm going to study the eigenvalues and eigenvector of the correlation matrix to see if any of those values hover around 0. 


```python
w,v = np.linalg.eig(cm)
print(w)
print(v[:,1])
```

Looks like the second eigen-value is the lowest (closest to 0). Looks like RM & ZN have multi-collinearity. Let's analyze while discluding ZN from the analysis and introduce it later if we need to. 

```python
imp_features = imp_features.drop(['ZN'],axis=1)
```

Now, let's apply a linear regression model on the unscaled data first and then we will try it with Standard Scaler to see how it impacts the model performance. 

```python
# Let's perform a train/test/split again
y = np.log(df_Boston['Price'])
x_train1,x_test1,y_train1,y_test1 = train_test_split(imp_features,y,test_size=0.33, random_state=22)
clf.fit(x_train1,y_train1)
pred = clf.predict(x_test1)
func = np.vectorize(lambda x:math.exp(x))
pred = func(pred)
error = mean_squared_error(func(y_test1),pred)
print(error)
```

Wow! this brought down the error from 34 all the way to 11 by including more features. Is this for real? Not sure, but let's move and try Standard Scaling and then doing the same thing. 

```python
# Let's bring in my Pipeline knowledge here and try to use the same procedure
steps = [('scaler', StandardScaler()),('linear', LinearRegression())]
pipeline = Pipeline(steps)
pipeline.fit(x_train1,y_train1)
scaled_predict = pipeline.predict(x_test1)
scaled_predict = func(scaled_predict)
scaled_error = mean_squared_error(func(y_test1),scaled_predict)
print(error)
```

Regularization is affected by scaling. Let's perform Ridge regression on unscaled vs. scaled and see how the model performs. 

```python
model = Ridge(alpha=1.0)
model.fit(x_train1,y_train1)
ridge_pred_unscale = model.predict(x_test1)
ridge_pred_unscale = func(ridge_pred_unscale)
ridge_error_unscale = mean_squared_error(func(y_test1),ridge_pred_unscale)
print(ridge_error_unscale)
```

```python
# Nothing changed, let's try applying standardscaler and trying
steps =  [('scaler', StandardScaler()),('ridge', Ridge(alpha=1.0))]
pipeline = Pipeline(steps)
pipeline.fit(x_train1,y_train1)
ridge_pred_scale = pipeline.predict(x_test1)
ridge_pred_scale = func(ridge_pred_scale)
ridge_error_scale = mean_squared_error(func(y_test1),ridge_pred_scale)
print(ridge_error_scale)
```

It looks like it did a better job !! Let's try a MinMaxScaler and see if that changes things

```python
steps =  [('scaler', MinMaxScaler()),('ridge', Ridge(alpha=1.0))]
pipeline = Pipeline(steps)
pipeline.fit(x_train1,y_train1)
ridge_pred_scale = pipeline.predict(x_test1)
ridge_pred_scale = func(ridge_pred_scale)
ridge_error_scale = mean_squared_error(func(y_test1),ridge_pred_scale)
score = r2_score(func(y_test),ridge_pred_scale)
print(ridge_error_scale)
print(score)
```
